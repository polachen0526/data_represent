```python
def greedy(pred):
    pred_result = np.argmax(pred,axis=-1)
    new_labels = []
    previous = None
    for l in pred_result:
        if l != previous:
            new_labels.append(l)
            previous = l
    new_labels = [l for l in new_labels if int(l) < len(char_list)]
    return new_labels
```

```python
Dense_float_result , Dense_fixed_result= Dense(units=63,file_number=1,float_layer=float_lstm_answer2,fixed_layer=fixed_lstm_answer2,weight_shape=(128,63),bias_shape=(1,63),relu_type="relu")
Dense_float_answer = np.argmax(Dense_float_result,axis=-1)
Dense_fixed_answer = np.argmax(Dense_fixed_result,axis=-1)
print("\nfloat每一個cell特徵值")
print(Dense_float_result)
print("\nfixed每一個cell特徵值")
print(Dense_fixed_result)
print("\nfloat每一個cell特徵最大值index")
print(Dense_float_answer)
print("\nfixed每一個cell特徵最大值index")
print(Dense_fixed_answer)
char_list = string.ascii_letters + string.digits
print("\n-------------------prediction-----------------\n")
print(greedy(Dense_float_result))
output = greedy(Dense_float_result)
for x in output:
    print("float_predicted_text =  ", end = '')
    if int(x) < 62 and int(x)!=-1:
        print(char_list[int(x)], end = '')       
    print('\n')
print("--------------------------------------")
print(greedy(Dense_fixed_result))
output = greedy(Dense_fixed_result)
for x in output:
    print("fixed_predicted_text =  ", end = '')
    if int(x) < 62 and int(x)!=-1:
        print(char_list[int(x)], end = '')       
    print('\n')
```
